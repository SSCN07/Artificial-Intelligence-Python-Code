```python
X_mean=np.mean(X)
X_n=X-X_mean
X_std=np.std(X_n,0)
X_norm=X_n/X_std
X_norm=X_norm.T
X_norm
```

```python
y
```

```python
y=y.T
y
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def tanh(z):
    return np.tanh(z)
```

```python
def init_param(X,y,hidden):
    n_x=X.shape[0]
    n_y=y.shape[0]
    w1=np.random.rand(hidden,n_x)-0.5
    b1=np.random.rand(hidden,1)-0.5
    w2=np.random.rand(n_y,hidden)-0.5
    b2=np.random.rand(n_y,1)-0.5
    return w1,b1,w2,b2
```

```python
params=init_param(X_norm,y,4)
w1,b1,w2,b2=params[0],params[1],params[2],params[3]
w1.shape
```

```python
b2.shape
```

```python
def forward_propa(X,y,params):
    w1=params[0]
    b1=params[1]
    w2=params[2]
    b2=params[3]

    Z1=w1.dot(X) + b1
    A1=tanh(Z1)
    Z2=w2.dot(A1) + b2
    A2=sigmoid(Z2)

    epsilon = 1e-15  # Small value to prevent log(0)
    # Clip predictions to avoid 0 and 1
    A2 = np.clip(A2, epsilon, 1 - epsilon)

    return Z1,A1,Z2,A2
```

```python
cache=forward_propa(X_norm,y,params)
Z1,A1,Z2,A2=cache[0],cache[1],cache[2],cache[3]
```

```python
A2.shape
```

```python
def costfunc(y,A2):
    m=y.shape[1]
    cost=-(1/m)*np.sum(y*np.log(A2)+(1-y)*(1-A2))
    return cost
```

```python
def back_propa(X,y,params,cache):
    w1=params[0]
    b1=params[1]
    w2=params[2]
    b2=params[3]

    Z1=cache[0]
    A1=cache[1]
    Z2=cache[2]
    A2=cache[3]

    m=y.shape[1]

    dZ2= A2 - y
    dw2=(1/m)*(dZ2.dot(A1.T))
    db2=(1/m)*np.sum(dZ2)

    dZ1= w2.T.dot(dZ2)*(1-A1**2)
    dw1=(1/m)*(dZ1.dot(X.T))
    db1=(1/m)*np.sum(dZ1)

    return dw2,db2,dw1,db1
```

```python
grads=back_propa(X_norm,y,params,cache)
dw2,db2,dw1,db1=grads[0],grads[1],grads[2],grads[3]
```

```python
db1.shape
```

```python
def update(grads,params,learn_r):
    w1=params[0]
    b1=params[1]
    w2=params[2]
    b2=params[3]

    dw2=grads[0]
    db2=grads[1]
    dw1=grads[2]
    db1=grads[3]

    w1=w1-learn_r*dw1
    b1=b1-learn_r*db1
    w2=w2-learn_r*dw2
    b2=b2-learn_r*db2

    return w1,b1,w2,b2
    
```

```python
params=update(grads,params,learn_r=0.8)
w1,b1,w2,b2=params
```

```python
def get_predictions(A2):
    return np.where(A2 >= 0.5, 1, 0)
```

```python
def get_accuracy(y_predicted, y):
    #print(y_predicted, y)
    return np.sum(y_predicted == y) / y.size
```

```python
def train(X,y,hidden=4,epoch=400000,learn_r=0.8):
    params=params=init_param(X_norm,y,hidden)
    cost_history=[]
    n_print=epoch//20
    for i in range(epoch):
        cache=forward_propa(X_norm,y,params)
        A2=cache[3]
        cost= costfunc(y,A2)
        grads=back_propa(X_norm,y,params,cache)
        params=update(grads,params,learn_r)
        cost_history.append(cost)

        if (i+1) % n_print == 0:
            predictions = get_predictions(A2)
            accu = get_accuracy(predictions, y)
            
            print("Iteration", i+1, " | Cost: ", cost, " | Accuracy: ", accu,"\n")

    w1=params[0]
    b1=params[1]
    w2=params[2]
    b2=params[3]

    return w1, b1, w2, b2, cost_history
            
        
    
```

```python
w1, b1, w2, b2, cost_history=train(X,y,hidden=4,epoch=4000000,learn_r=0.8)
```
